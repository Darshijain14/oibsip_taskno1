```python
# Plot scatterplots to visualize relationships between features
colors = ['yellow', 'green', 'red']
species = ['Iris-setosa', 'Iris-versicolor' , 'Iris-virginica']
```

visualization using matplotlib

```python
# Scatter plot comparing Sepal Length and Width by species
for idx in range(3):
    subset = df[df['Species'] == species[idx]]
    plt.scatter(subset['SepalLengthCm'], subset['SepalWidthCm'], color=colors[idx], label=species[idx])

plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Sepal Dimensions by Species")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Visualizing Petal Length vs Width across species
for idx in range(3):
    data_subset = df[df['Species'] == species[idx]]
    plt.scatter(data_subset['PetalLengthCm'], data_subset['PetalWidthCm'],
                color=colors[idx], label=species[idx], edgecolor='black')

plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.title("Petal Size Distribution by Species")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting Sepal vs Petal Length grouped by species
for idx in range(3):
    group = df[df['Species'] == species[idx]]
    plt.scatter(group['SepalLengthCm'], group['PetalLengthCm'],
                color=colors[idx], label=species[idx], alpha=0.7, edgecolor='black')

plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.title("Sepal vs Petal Length by Species")
plt.legend()
plt.grid(True)
plt.show()
```

### **Preparing the Data and Building the Model**

```python
# Convert categorical class labels to numerical values using Label Encoding
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Species'] = encoder.fit_transform(df['Species'])
df['Species']
```

```python
# check for null values
df.isnull().sum()
```

```python
# Importing the function to split data into training and testing sets
from sklearn.model_selection import train_test_split

# Defining the feature set and target variable
features = df.drop(columns=['Species'])
target = df['Species']

# Splitting the data into training (60%) and testing (40%) sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.40, random_state=42)

```

```python
# Applying standard scaling to normalize the feature data
from sklearn.preprocessing import StandardScaler

standardizer = StandardScaler()
normalized_features = standardizer.fit_transform(features)

print(features)
```

### **Model Training**

```python
# K-nearest Neighbours Model (KNN)
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X_train.values, y_train.values) #Here x.values will have only values without headers
# X = X.values # conversion of X  into array
# print metric to get performance
print("Accuracy (KNN): ", model.score(X_test.values, y_test.values) * 100)
```
